```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd

# Fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# Data (as pandas dataframes) 
# X contains the features (age, sex, etc.) and y contains the target (disease presence)
X = heart_disease.data.features 
y = heart_disease.data.targets 

# Combine them into one "dataframe" so it looks like a normal table
df = pd.concat([X, y], axis=1)

# Display the first 5 rows to make sure it worked
print(df.head())

# SAVE AS CSV (So you can give it to your teammates and upload to GitHub)
df.to_csv('heart_disease.csv', index=False)
```
